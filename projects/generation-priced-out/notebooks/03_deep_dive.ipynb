findings summary.

**Available analysis types:**
- Regression (does A predict B after controlling for confounders?)
- Lag analysis (does X precede Y by N years?)
- Segmented comparison (high-X vs low-X regions on outcome)
- Before/after analysis (metrics before vs after an event year)
- Narrative pattern test (masking, displacement, divergence, etc.)

```python
from pathlib import Path

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from pipeline.config import load_config, get_data_processed_dir

PROJECT_NAME = "qol-immigration"  # UPDATE per project

config = load_config(PROJECT_NAME)
processed_dir = get_data_processed_dir(config)

parquet_files = sorted(processed_dir.glob("*.parquet"))
datasets = {pf.stem: pd.read_parquet(pf) for pf in parquet_files}
print(f"Loaded {len(datasets)} dataset(s)")
```

## Finding Details

Update these variables to match the finding you're investigating.

```python
# --- UPDATE THESE PER DEEP-DIVE ---
FINDING_NAME = "Example finding"  # Short name from findings summary
ANALYSIS_TYPE = "regression"       # regression | lag | segmented | before_after | pattern
PATTERN_NAME = None                # If ANALYSIS_TYPE is 'pattern': masking | displacement | divergence | lag_lead | segmentation | threshold

# Variables involved
OUTCOME_VAR = ""       # The dependent / outcome variable
PREDICTOR_VAR = ""     # The main predictor
CONTROL_VARS = []      # Confounders to control for (regression)
SEGMENT_VAR = ""       # Variable to segment by (segmented comparison)
EVENT_YEAR = None       # Year of event/policy change (before/after)
HIDDEN_VAR = ""        # Masking variable (pattern: masking)
TIME_COL = "year"
GEO_COL = "state"
```

## Regression Analysis

```python
if ANALYSIS_TYPE == "regression":
    from pipeline.patterns import run_regression
    results = run_regression(datasets, OUTCOME_VAR, PREDICTOR_VAR,
                             controls=CONTROL_VARS)
    print(results["summary"])
    if results.get("figure"):
        results["figure"].show()
else:
    print("Skipping regression (not selected).")
```

## Lag Analysis

```python
if ANALYSIS_TYPE == "lag":
    from pipeline.patterns import run_lag_analysis
    results = run_lag_analysis(datasets, PREDICTOR_VAR, OUTCOME_VAR,
                               time_col=TIME_COL, max_lag=10)
    print(results["summary"])
    if results.get("figure"):
        results["figure"].show()
else:
    print("Skipping lag analysis (not selected).")
```

## Segmented Comparison

```python
if ANALYSIS_TYPE == "segmented":
    from pipeline.patterns import run_segmented_comparison
    results = run_segmented_comparison(datasets, PREDICTOR_VAR, OUTCOME_VAR,
                                       segment_var=SEGMENT_VAR)
    print(results["summary"])
    for fig in results.get("figures", []):
        fig.show()
else:
    print("Skipping segmented comparison (not selected).")
```

## Before/After Analysis

```python
if ANALYSIS_TYPE == "before_after":
    from pipeline.patterns import run_before_after
    results = run_before_after(datasets, OUTCOME_VAR,
                                event_year=EVENT_YEAR, time_col=TIME_COL)
    print(results["summary"])
    if results.get("figure"):
        results["figure"].show()
else:
    print("Skipping before/after analysis (not selected).")
```

## Narrative Pattern Test

```python
if ANALYSIS_TYPE == "pattern" and PATTERN_NAME:
    from pipeline.patterns import run_pattern
    results = run_pattern(
        datasets,
        pattern=PATTERN_NAME,
        outcome_var=OUTCOME_VAR,
        predictor_var=PREDICTOR_VAR,
        hidden_var=HIDDEN_VAR,
        segment_var=SEGMENT_VAR,
        time_col=TIME_COL,
        event_year=EVENT_YEAR,
    )
    print(results["summary"])
    for fig in results.get("figures", []):
        fig.show()
else:
    print("Skipping pattern test (not selected or no pattern specified).")
```
